```r
library(rstan)
library(bayesplot)
```

```r
head(cars, n=3)
```

```r
sample_size <- nrow(cars)
sample_size
```

```r
plot(cars, main = "cars")
```

```r
speed_pred <- 5:25
speed_pred
```

```r
data_list_pred<- list(
    N=sample_size,
    dist=cars$dist,
    speed=cars$speed,
    N_pred = length(speed_pred),
    speed_pred=speed_pred
)
```

```r
mcmc_result <- stan(
    file="simple_lm-vec.stan",
    data=data_list_pred,
    seed=1
)
```

```r
print(mcmc_result, probs=c(0.025, 0.5, 0.975))
```

```r
mcmc_sample <- rstan::extract(mcmc_result, permuted=FALSE)
```

```r
mcmc_combo(
    mcmc_sample,
    pars = c("Intercept" , "beta", "sigma")
)
```

```r
mcmc_sample_pred <- rstan::extract(mcmc_result, permuted=FALSE)
```

```r
mcmc_intervals(
    mcmc_sample_pred,
    regex_pars = c("dist_pred."),
    prob = 0.8,
    prob_outer = 0.95
)
```

```r
mcmc_intervals(
    mcmc_sample_pred,
    pars = c("mu_pred[1]", "dist_pred[1]"),
    prob = 0.8,
    prob_outer = 0.95
)
```

```r
mcmc_areas(
    mcmc_sample_pred,
    pars = c("dist_pred[1]", "dist_pred[21]"),
    prob = 0.6,
    prob_outer = 0.99
)
```
